```python
import joblib
import pandas as pd
import gradio as gr
from google.colab import drive

# Cargar el modelo desde la ruta de Drive
ruta_drive = '/content/drive/MyDrive/Inteligencia Computacional/modelo_diabetes_final.pkl'
modelo_cargado = joblib.load(ruta_drive)

# Funci√≥n de predicci√≥n (Inferencia)
def predecir_diabetes(bp, chol, bmi, smoker, alcohol, health, age):
    # Crear el DataFrame con los nombres de columnas que el modelo conoce
    df_input = pd.DataFrame([[
        1.0 if bp == "S√≠" else 0.0,
        1.0 if chol == "S√≠" else 0.0,
        float(bmi),
        1.0 if smoker == "S√≠" else 0.0,
        1.0 if alcohol == "S√≠" else 0.0,
        float(health),
        float(age)
    ]], columns=['HighBP', 'HighChol', 'BMI', 'Smoker', 'HvyAlcoholConsump', 'GenHlth', 'Age'])

    # Realizar la predicci√≥n
    resultado = modelo_cargado.predict(df_input)[0]

    return "üö© RIESGO DE DIABETES DETECTADO" if resultado == 1 else "‚úÖ RESULTADO NORMAL / SANO"

# configuraci√≥n de la Interfaz Gr√°fica (GUI)
interfaz = gr.Interface(
    fn=predecir_diabetes,
    inputs=[
        gr.Radio(["S√≠", "No"], label="¬øTiene Presi√≥n Alta? (HighBP)", value="No"),
        gr.Radio(["S√≠", "No"], label="¬øTiene Colesterol Alto? (HighChol)", value="No"),
        gr.Slider(10, 60, step=0.1, label="√çndice de Masa Corporal (BMI)", value=25.0),
        gr.Radio(["S√≠", "No"], label="¬øEs fumador? (Smoker)", value="No"),
        gr.Radio(["S√≠", "No"], label="¬øConsume alcohol en exceso? (HvyAlcohol)", value="No"),
        gr.Slider(1, 5, step=1, label="Estado de Salud Percibido (1:Excelente - 5:Mala)", value=2),
        gr.Slider(1, 13, step=1, label="Rango de Edad (1: 18-24 hasta 13: 80+)", value=5)
    ],
    outputs=gr.Textbox(label="Resultado del Diagn√≥stico IA"),
    title="ü©∫ Asistente Inteligente de Salud: Riesgo de Diabetes",
    description="""Esta aplicaci√≥n utiliza un modelo de Machine Learning (Random Forest) entrenado con datos de salud de m√°s de 70,000 pacientes.
    Complete los datos para obtener una evaluaci√≥n preliminar.""",
    theme="soft" # Un tema visualmente m√°s limpio
)

# Lanzar la aplicaci√≥n
# share=True genera un link p√∫blico que puedes enviarle a tu profesor
interfaz.launch(share=True)
```
