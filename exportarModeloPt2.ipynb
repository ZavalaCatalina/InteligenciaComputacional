```python
import joblib
import pandas as pd
from google.colab import drive

# 1. Montar Drive y Cargar el Modelo
drive.mount('/content/drive')
ruta_drive = '/content/drive/MyDrive/Inteligencia Computacional/modelo_diabetes_final.pkl'
modelo_cargado = joblib.load(ruta_drive)

#Solicita y validar inputs del usuario(dinamico)
def solicitar_dato(mensaje, min_v, max_v, es_bool=False):

    while True:
        try:
            texto = input(mensaje).strip().upper()
            if es_bool:
                if texto in ['S', 'N']:
                    return 1.0 if texto == 'S' else 0.0
                print(" Por favor, responda con S o N.")
            else:
                valor = float(texto)
                if min_v <= valor <= max_v:
                    return valor
                print(f" Error: El nÃºmero debe estar entre {min_v} y {max_v}.")
        except ValueError:
            print(" Error: Ingrese un valor numÃ©rico vÃ¡lido.")
#Programa

def ejecutar_proyecto():
    columnas = ['HighBP', 'HighChol', 'BMI', 'Smoker', 'HvyAlcoholConsump', 'GenHlth', 'Age']

    # PREDICCIÃ“N HARDCODEADA
    print("\n--- EJECUTANDO PREDICCIÃ“N HARDCODEADA (TEST) ---")
    datos_hardcoded = [[1.0, 1.0, 32.0, 1.0, 0.0, 4.0, 9.0]] # Paciente ficticio
    df_hard = pd.DataFrame(datos_hardcoded, columns=columnas)
    test_pred = modelo_cargado.predict(df_hard)[0]
    print(f"Resultado para datos predefinidos: {'DIABETES' if test_pred == 1 else 'SANO'}")
    print("-" * 50)

    # INTERFAZ DINÃMICA
    print("\nðŸ©º ASISTENTE DE SALUD INTERACTIVO")

    bp      = solicitar_dato("1. Â¿Tiene PresiÃ³n Alta? (S/N): ", 0, 0, es_bool=True)
    chol    = solicitar_dato("2. Â¿Tiene Colesterol Alto? (S/N): ", 0, 0, es_bool=True)
    bmi     = solicitar_dato("3. Ingrese su IMC/BMI (Ejemplo: 25.5): ", 10, 60)
    smoker  = solicitar_dato("4. Â¿Es fumador? (S/N): ", 0, 0, es_bool=True)
    alcohol = solicitar_dato("5. Â¿Consume alcohol en exceso? (S/N): ", 0, 0, es_bool=True)
    health  = solicitar_dato("6. Salud percibida (1:Excelente a 5:Mala): ", 1, 5)

    # GuÃ­a visual para la Edad
    print("\n--- GUÃA DE RANGOS DE EDAD ---")
    print("1: 18-24 | 2: 25-29 | 3: 30-34 | 4: 35-39 | 5: 40-44 | 6: 45-49")
    print("7: 50-54 | 8: 55-59 | 9: 60-64 | 10: 65-69 | 11: 70-74 | 12: 75-79 | 13: 80+")

    age = solicitar_dato("7. Seleccione el nÃºmero de su rango (1-13): ", 1, 13)

    # Inferencia con datos del usuario
    df_usuario = pd.DataFrame([[bp, chol, bmi, smoker, alcohol, health, age]], columns=columnas)
    prediccion = modelo_cargado.predict(df_usuario)[0]

    print("\n" + "="*40)
    if prediccion == 1:
        print(" RESULTADO: RIESGO DE DIABETES DETECTADO")
    else:
        print("RESULTADO: NORMAL / SANO")
    print("="*40)

if __name__ == "__main__":
    ejecutar_proyecto()
```
